```python
def make(x):
    times=time.mktime(datetime.datetime.strptime(x, "%Y-%m-%d %X").timetuple())
    return times

data['make']=data.date.apply(make)
x = data[['x','y','date', '']]
new_x = x[(data.x < -121)]
print len(new_x)
print len(x)
```

```python
# to make inertia curve

def make_inertia():
    krange = range(1, 20)
sum_squares = [cluster.KMeans(n_clusters=k).fit(new_x).inertia_ for k in krange]
variance_explained = [1.0 - (s / sum(sum_squares)) for s in sum_squares]
plt.figure()
plt.plot(range(1,20), variance_explained)
plt.ylabel('% of explained variance')
plt.xlabel('clusters K')
plt.xlim([1, 20])
```
